# Simple Scatter Plots

__Import Libraries__

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

---

## Scatter Plots with plt.plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black'); # third argument is for marker keyword
```

```python
rng = np.random.RandomState(21)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
    plt.plot
```

```python
plt.plot(x, y, '-ok'); # line(-), circle marker(o), black(k)
```

```python
plt.plot(x, y, '-o', color='gray',
        markersize=7, linewidth=1,
        markerfacecolor='cyan',
        markeredgecolor='gray',
        markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

---

## Scatter Plots with plt.scatter

```python
plt.scatter(x, y, marker='o')
plt.scatter(x, np.cos(x), marker='+')
```

The primary difference of plt.scatter from plt.plot is that it can be used to create scatter plots where the properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data.

```python
# aplha for transparency level
rng = np.random.RandomState(21)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=color, s=sizes, alpha=0.3,
           cmap='viridis')
plt.colorbar(); # show color scale
```

Notice that the color argument is automatically mapped to a color scale (shown here by the colorbar() command), and the size argument is given in pixels. In this way, the color and size of points can be used to convey information in the visualization, in order to illustrate multidimensional data.

```python
# iris dataset from sklearn
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

For large datasets, the difference between these two can lead to vastly different performance, and for this reason, plt.plot should be preferred over plt.scatter for large datasets.

## Experimentation

```python
plt.plot(rng.randn(100), rng.randn(100), 'om');
```

```python
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(rng.randn(100), rng.randn(100), c=colors, s=sizes,
           alpha=0.25, cmap='inferno')
plt.colorbar();
```

```python
colors = np.linspace(0, 1, 100)
sizes = np.linspace(1, 0, 100) * 500

x = np.linspace(0, 100, 100)
y = np.sin(x)

plt.scatter(x, y, s=sizes, c=colors,
           alpha=0.5, cmap='viridis')
plt.colorbar();
```
