# Customizing Ticks

__Import libraries__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
ax = plt.axes(xscale='log', yscale='log')
```

```python
print(ax.xaxis.get_major_locator())
print(ax.xaxis.get_minor_locator())
```

```python
print(ax.xaxis.get_major_formatter())
print(ax.xaxis.get_minor_formatter())
```

We see that both major and minor tick labels have their locations specified by a LogLocator (which makes sense for a logarithmic plot). Minor ticks, though, have their labels formatted by a NullFormatter; this says that no labels will be shown.

---

## Hiding Ticks or Labels

```python
ax = plt.axes()
ax.plot(np.random.rand(50))

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
ax = plt.axes()
ax.scatter(np.random.randn(100), np.random.randn(100));

ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.yaxis.set_major_formatter(plt.NullFormatter())
# locator is a line grid and formtter is its label ticks
# if we are removing the locator, we also removing the ticks label
```

## Reducing or Increasing the Number of Ticks

```python
# default will end up with crowded labels
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True)
```

```python
# for every axis, set the x and y major locator
for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
fig
```

## Fancy Tick Formats

```python
# plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')

# set up grid, legend and limits
ax.grid(True)
ax.legend(frameon=False)
ax.axis(aspect='equal')
ax.set_xlim(0, 3 * np.pi);
```

```python
# setting MultipleLocator
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
fig
```

```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```

## Experimentation

```python
fig, ax = plt.subplots()
x = np.linspace(-(3 * np.pi), 3 * np.pi)
y = np.sin(x)
ax.plot(x, y, '-.b')
ax.set_title('Sine function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.yaxis.set_major_locator(plt.MaxNLocator(5))
```
