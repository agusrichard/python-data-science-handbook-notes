## Two-Dimensional Histograms and Binnings

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
print(x.shape)
print(y.shape)
```

### plt.hist2d: Two-dimensional histogram

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin');
```

```python
# counter-part of np.histogram in two-dimension
counts, xedges, yedges = np.histogram2d(x, y, bins=3)
print(counts) # numpy array 3x3
# for dimension higher than two, use np.histogramdd function
```

### plt.hexbin: Hexagonal binnings

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin');
```

plt.hexbin has a number of interesting options, including the ability to specify weights for each point, and to change the output in each bin to any NumPy aggregate (mean of weights, standard deviation of weights, etc.).

### Kernel density estimation

KDE can be thought of as a way to “smear out” the points in space and add up the result to obtain a smooth function.

```python
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 100)
ygrid = np.linspace(-6, 6, 100)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
origin='lower', aspect='auto',
extent=[-3.5, 3.5, -6, 6],
cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```

KDE has a smoothing length that effectively slides the knob between detail and
smoothness (one example of the ubiquitous bias–variance trade-off). The literature on choosing an appropriate smoothing length is vast: gaussian_kde uses a rule of thumb to attempt to find a nearly optimal smoothing length for the input data.

## Experimentation

```python
data = np.random.randn(10000)
```

```python
plt.hist(data);
```

```python
plt.hist(data, bins=50);
```

```python
plt.hist(data, bins=50, alpha=0.75, normed=True);
```

```python
x = np.random.chisquare(21, 10000)
y = np.random.normal(10, 2.5, 10000)
z = np.random.exponential(10, size=10000)
```

```python
kwargs = dict(bins=100, density=True, alpha=0.5)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
plt.hist(z, **kwargs)
plt.axis([0, 50, 0, 0.20]);
```

```python
x = np.random.normal(0, 5, size=10000)
y = np.random.normal(0, 5, size=10000)

plt.hist2d(x, y, alpha=0.75, cmap='viridis', bins=100)
plt.colorbar(label='Density');
```

```python
def f(x):
    return np.sin(x)

x1 = np.random.normal(size=1000)
y1 = f(x1)

plt.hist2d(x1, y1, bins=50, alpha=0.7, cmap='viridis');
```
