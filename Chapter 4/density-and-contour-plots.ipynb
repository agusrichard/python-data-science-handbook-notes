# Density and Contour Plots

__Import libraries__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-white')
```

---

## Visualizing a Three-Dimensional Function

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

A contour plot can be created with the plt.contour function. It takes three arguments: a grid of x values, a grid of y values, and a grid of z values. The x and y values represent positions on the plot, and the z values will be represented by the contour levels. Perhaps the most straightforward way to prepare such data is to use the np.meshgrid function, which builds two-dimensional grids from one-dimensional arrays:

```python
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

Notice that by default when a single color is used, negative values are represented by dashed lines, and positive values by solid lines.

```python
plt.contour(X, Y, Z, cmap='inferno');
```

```python
# filled contour plot
plt.contourf(X, Y, Z, cmap='RdGy')
plt.colorbar();
```

Use the plt.imshow() function, which interprets a two-dimensional grid of data as an image.

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
          cmap='RdGy') # extent specify x-y lim
plt.colorbar()
plt.axis(aspect='image'); # adjust aspect ratio, so x and y units match
```

```python
# transparency set by alpha parameter
# overplot contours with labels on the contours themselves
contours = plt.contour(X, Y, Z, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
          cmap='RdGy', alpha=0.5)
plt.colorbar();
```

The combination of these three functions—plt.contour, plt.contourf, and
plt.imshow—gives nearly limitless possibilities for displaying this sort of threedimensional data within a two-dimensional plot.

## Experimentation

```python
def func1(x, y):
    return np.sin(x) + np.cos(y)
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = func1(X, Y)
print(Z.shape)
```

```python
plt.contour(X, Y, Z, colors='blue');
```

```python
plt.contour(X, Y, Z, cmap='viridis');
```

```python
plt.contourf(X, Y, Z, cmap='inferno')
plt.colorbar();
```

```python
plt.imshow(Z, extent=[-10, 10, -10, 10], origin='lower', cmap='viridis')
plt.colorbar()
plt.axis(aspect='image');
```

```python
contours = plt.contour(X, Y, Z, colors='black')
plt.clabel(contours, inline=True, fontsize=5)

plt.imshow(Z, extent=[-10, 10, -10, 10], origin='lower', cmap='viridis', alpha=0.9)
plt.colorbar();
```
